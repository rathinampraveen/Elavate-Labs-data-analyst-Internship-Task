```python
import pandas as pd
import sqlite3
```

```python
# Load your CSV file
csv_file = r"C:\Users\MMM\Downloads\ecommerce-consumer-behavior-analysis-data\Ecommerce_Consumer_Behavior_Analysis_Data.csv"
df = pd.read_csv(csv_file)

# Create a connection to the SQLite database (creates file if it doesn't exist)
conn = sqlite3.connect('your_database.db')

# Write the data to a table (it will create the table if it doesn't exist)
df.to_sql('your_table_name', conn, if_exists='replace', index=False)

# Close the connection
conn.close()

print("CSV has been successfully converted to SQL database.")
```
